```python
## Counter plot 확인
from collections import Counter
print('행정동:', Counter(data_dong["서비스_업종_코드_명"]))
print('상권:', Counter(data_market["서비스_업종_코드_명"]))
```

```python
def plot_count(data, var, title='행정동'):
    # Counter를 사용하여 각 항목의 빈도를 계산
    counter_data = Counter(data[var])
    # sort values
    sorted_data = dict(sorted(counter_data.items(), key=lambda x: x[1], reverse=True))

    # 바 차트 그리기
    categories = list(sorted_data.keys())
    counts = list(sorted_data.values())
    plt.figure(figsize=(20,10))
    plt.bar(categories, counts)
    plt.xlabel(var, fontsize=15)
    plt.ylabel('빈도', fontsize=15)
    plt.title(var+' 빈도수-'+title , fontsize=20)
    plt.xticks(rotation=90)
    plt.show()
```

```python
plot_count(data_dong, "행정동_코드_명")
plot_count(data_dong, "서비스_업종_코드_명")
plot_count(data_market, "상권_코드_명", '상권')
plot_count(data_market, "서비스_업종_코드_명", '상권')
plot_count(data_market, "상권_구분_코드_명", '상권')
```

### 매출 분포 확인

```python
import seaborn as sns
## 매출 분포 확인
def plot_sales(data=data_dong, var="행정동_코드_명", title='행정동'):
    # 각 동/상권별 매출 합계 구하기
    whole_sales = data.groupby([var])["당월_매출_금액"].sum()
    
    # 각 동/상권별 분기별 매출 합계 구하기
    quarter_sales = data.groupby([var, "기준_년분기_코드"])["당월_매출_금액"].sum()
    
    # sort values
    sorted_whole = dict(sorted(whole_sales.items(), key=lambda x: x[1], reverse=True))

    # 바 차트 그리기
    categories = list(sorted_whole.keys())
    counts = list(sorted_whole.values())
    
    plt.figure(figsize=(30,10))
    plt.bar(categories, counts)
    plt.xlabel(var, fontsize=15)
    plt.ylabel('총 매출', fontsize=15)
    plt.title(var+' 총 매출-'+title , fontsize=20)
    # 라벨이 겹치지 않도록 조정
    plt.tight_layout()
    plt.xticks(rotation=90)
 
    plt.show()
    
    plt.bar(quarter_sales)
    
    
```
