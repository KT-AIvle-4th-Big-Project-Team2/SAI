```python
# vif 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# X null 값 drop 진행 -> 2020-1~2022-4까지 존재
X_fin = X.dropna()

vif = pd.DataFrame()
scaler = StandardScaler()

train_x_scaled = scaler.fit_transform(X_fin)
vif["feature"] = X_fin.columns
vif["VIF"] = [variance_inflation_factor(train_x_scaled, i) for i in range(len(data.columns))]

# vif 높은 순으로 데이터 정렬
vif.sort_values(by = 'VIF')
```

```python
# vif 10이상인 변수만 확인
vif[vif["VIF"]>=10].sort_values(by = 'VIF')
```

### 상관계수 확인

```python
def num_plot(df, var):
    plt.figure(figsize=(16, 10))
    plt.subplot(1, 2, 1)
    sns.histplot(df[var], kde=True)
    plt.axvline(df[var].mean(), color='red', linestyle='dashed', linewidth=2)
    plt.text(df[var].mean() + df[var].mean() / 10, 100, 'mean: {:.2f}'.format(df[var].mean()), fontsize=13)
    plt.title(var + ' histogram')

    plt.subplot(1, 2, 2)
    sns.boxplot(df[var])
    plt.title(var + ' boxplot')
    plt.show()
```
