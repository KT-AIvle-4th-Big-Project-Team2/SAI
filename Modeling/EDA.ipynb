```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')

import warnings
warnings.filterwarnings('ignore')
```

## 상권

### 상주인구

```python
상주인구 = pd.read_csv('../../data/raw data/상권/상주인구/서울시 상권분석서비스(상주인구-상권).csv', encoding='cp949')
상주인구
```

```python
상주인구.info()
```

```python
상주인구 = 상주인구.drop(columns=['상권_구분_코드', '상권_코드'])
상주인구
```

```python
sns.barplot(x='상권_구분_코드_명', y='총_상주인구_수', data=상주인구, ci=None, estimator='mean')
plt.title('상권별 상주인구(평균)')
plt.show()
```

```python
발달상권 = 상주인구[상주인구['상권_구분_코드_명'] == '발달상권']
발달상권 = 발달상권.drop(columns=['기준_년분기_코드', '상권_구분_코드_명', '상권_코드_명'])

plt.figure(figsize=(20,6))
sns.barplot(발달상권, ci=None)
plt.title('발달상권 상주인구(평균)')
plt.xticks(rotation=90)
plt.show()
```

```python
전통시장 = 상주인구[상주인구['상권_구분_코드_명'] == '전통시장']
전통시장 = 전통시장.drop(columns=['기준_년분기_코드', '상권_구분_코드_명', '상권_코드_명'])

plt.figure(figsize=(20,6))
sns.barplot(전통시장, ci=None)
plt.title('전통시장 상주인구(평균)')
plt.xticks(rotation=90)
plt.show()
```
